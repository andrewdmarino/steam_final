```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import calendar
from datetime import datetime
from pandas import Series
from math import ceil
```

```python
steam_data = pd.read_csv('final_project_hits.csv')
```

```python
steam_data.head()
```

```python
cols = ['Developers', 'Publishers', 'Genres','Tags']

for col in cols:
    chart = steam_data[['Name', col]].groupby([col]).count().sort_values('Name', ascending=False).reset_index()
    sns.set_style("white")
    plt.figure(figsize=(12.4, 5))
    plt.xticks(rotation=90)
    sns.barplot(x=col, y='Name', data=chart[:30], palette=sns.cubehelix_palette((12 if col == 'clean_genres' else 30), dark=0.5, light=.90, reverse=True)).set_title(('Game count by '+col), fontsize=16)
    plt.ylabel('Count', fontsize=14)
    plt.xlabel('')
```

```python
heatmap2_data = pd.pivot_table(steam_data, values='Hit', 
                     index=['Genres'], 
                     columns='Metascores')
f, ax = plt.subplots(figsize=(14, 11))
heatmap_1 = sns.heatmap(heatmap2_data, cmap="YlGnBu")

figure_1 = heatmap_1.get_figure()    
figure_1.savefig('heatmap_1.jpg', dpi=400)
```

```python
heatmap3_data = pd.pivot_table(steam_data, values='Hit', 
                     index=['Tags'], 
                     columns='Metascores')

f, ax = plt.subplots(figsize=(17, 11))

heatmap_2= sns.heatmap(heatmap3_data, cmap="YlGnBu")
figure_2 = heatmap_2.get_figure()    
figure_2.savefig('heatmap_2.jpg', dpi=400)
```
